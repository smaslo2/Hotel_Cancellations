## Unsupervised Modeling
Unsupervised kmeans and nmf models were built on 85% of the training data.  Two clusters had the highest silhouette score, but contributed nothing to the model.  NMF with three clusters was slightly more effective at pulling useful information.  One hot encoding is redone here to remove the first result.

```python
df_onehot = pd.get_dummies(df, drop_first=True)
columns = df_onehot.columns
train_df = df_onehot.sample(frac=.85, replace=False, random_state=46)
train_df.shape
```

```python
from sklearn.decomposition import NMF
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(train_df.drop('booking_status', axis=1))
```

```python
scores = {'kmeans':[], 'NMF':[]}
for i in range(2,15):
    model = KMeans(n_clusters=i).fit(X)
    s = silhouette_score(X, model.labels_)
    scores['kmeans'].append(s)
    W = NMF(n_components = i, init='nndsvda', max_iter=400).fit_transform(X)
    s = silhouette_score(X, np.argmax(W, axis=1))
    scores['NMF'].append(s)
```

```python
plt.plot(range(2,15), scores['kmeans'], label='KMeans')
plt.plot(range(2,15), scores['NMF'], label='NMF')
plt.legend(loc='lower left')
plt.xlabel('Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs Clusters by Model')
```

```python
kmeans = KMeans(n_clusters=2).fit(X)
nmf_model = NMF(n_components=3, init='nndsvda')
nmf = nmf_model.fit_transform(X)
```

```python
train_df['nmf'] = np.argmax(nmf, axis=1)
train_df['kmeans'] = kmeans.labels_
```

## Model Training
Training a logistic regression, support vector machine, and gradient boosting classifier using 5-fold cross validation and accuracy as a metric.  Hyperparameter tuning was done for the C parameter in the SVC and number of stumps in the gradient boosted tree.  500 stumps provided nearly the best accuracy at a reasonable computation time.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from datetime import datetime
```

```python
train_df.head()
```

```python
train_df = train_df.drop('kmeans', axis = 1)
columns = train_df.columns.drop('booking_status')
cv_scaler = MinMaxScaler()
X = cv_scaler.fit_transform(train_df.drop('booking_status', axis=1))
```

```python
#using sag solver since liblinear struggled with convergence
LRmodel = LogisticRegression(solver='sag')
s = datetime.now()
LR_scores = cross_val_score(LRmodel, X=X, y=train_df.booking_status, cv=5)
LR_time = datetime.now() - s
print(LR_scores, round(np.mean(LR_scores),4), LR_time.total_seconds())
LR_score = np.mean(LR_scores)
LR_time = LR_time.total_seconds()
```

```python
SVC_C = [0.01, 0.1, .5, 1, 1.3]
SVC_scores = []
SVC_times = []
for C in SVC_C:
    SVCmodel= SVC(C=C, kernel='rbf')
    s = datetime.now()
    scores = cross_val_score(SVCmodel, X=X, y=train_df.booking_status, cv=5)
    timer = datetime.now()-s
    SVC_scores.append(np.mean(scores))
    SVC_times.append(timer.total_seconds())
    print(C, scores, round(np.mean(scores), 4), timer.total_seconds())
```

```python
gb_trees = [100,200,500,1000, 1200, 1500]
gb_scores = []
gb_times = []
for n in gb_trees:
    gbmodel = GradientBoostingClassifier(n_estimators = n, max_depth=1)
    s = datetime.now()
    scores = cross_val_score(gbmodel, X=X, y=train_df.booking_status, cv=5)
    timer = datetime.now()-s
    print(n, scores, round(np.mean(scores), 4), timer.total_seconds())
    gb_scores.append(np.mean(scores))
    gb_times.append(timer.total_seconds())
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, sharey='col', figsize=(12,6))
axs[0,0].plot(gb_trees, gb_scores, color='b')
axs[0,0].plot(gb_trees, [LR_score]*len(gb_trees), color='r')
axs[0,1].plot(gb_trees, gb_times, color='b')
axs[0,1].plot(gb_trees, [LR_time]*len(gb_trees), color='r')
axs[1,0].plot(SVC_C, SVC_scores, color='c')
axs[1,0].plot(SVC_C, [LR_score]*len(SVC_C), color='r')
axs[1,1].plot(SVC_C, SVC_times, color='c')
axs[1,1].plot(SVC_C, [LR_time]*len(SVC_times), color='r')
plt.setp(axs[:,0], ylabel='Accuracy')
plt.setp(axs[:,1], ylabel='Training Time')
plt.setp(axs[0,:], xlabel='Stumps')
plt.setp(axs[1,:], xlabel='Regularization')
fig.suptitle('Accuracy and Training Time for Models')
plt.show()
```

Finer hyperparameter tuning looking at the range of 500 to 1200 trees for gradient boosting classifier.  500 trees has nearly the same accuracy as the full data set, but takes less time to run.

```python
gb_trees2 = [500, 600, 700, 800, 900, 1000, 1100, 1200]
gb_scores2 = []
for n in gb_trees2:
    gbmodel = GradientBoostingClassifier(n_estimators = n, max_depth=1)
    scores = cross_val_score(gbmodel, X=X, y=train_df.booking_status, cv=5)
    gb_scores2.append(np.mean(scores))
plt.plot(gb_trees2, gb_scores2)
```

## Model Evaluation and Feature Importance
Compute training metrics on model fit to entire dataset.  The model will be built with 500 trees since gains beyond that are minimal and greatly contribute to computation time.

```python
training_model = GradientBoostingClassifier(n_estimators = 500, max_depth=1)
training_model.fit(X, train_df.booking_status)
training_predictions = training_model.predict(X)
training_proba = training_model.predict_proba(X)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
```

```python
train_metrics = {
    'split': 'Train',
    'roc_auc': roc_auc_score(train_df.booking_status, training_proba[:,1]),
    'recall': recall_score(train_df.booking_status, training_predictions),
    'precision': precision_score(train_df.booking_status, training_predictions),
    'accuracy': accuracy_score(train_df.booking_status, training_predictions)
}
train_confusion = confusion_matrix(train_df.booking_status, training_predictions)

for key, value in train_metrics.items():
    print(f'{key}: {value}')
print(train_confusion)
```

```python
from sklearn.inspection import permutation_importance
```

```python
importances = permutation_importance(training_model, X, train_df.booking_status)
feature_importance = {'Features':columns, 'Permuted':importances['importances_mean'], 'Model':training_model.feature_importances_}
feature_importance = pd.DataFrame(feature_importance)
#feature_importance
```

```python
plt.subplots(figsize=(6,6))
sns.barplot(y=feature_importance.Features, x=feature_importance.Permuted)
plt.show()
plt.subplots(figsize=(6,6))
sns.barplot(y=feature_importance.Features, x=feature_importance.Model)
plt.show()
```
